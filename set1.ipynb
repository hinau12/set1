```python
numbers = [12,45,7,89,23,5]

largest = max(numbers)
smallest = min(numbers)

print("largest number:", largest)
print("smallest number:",smallest)
```

```python
numbers = [10, 120, 90, 23, 67, 59, 191, 234]

even_count = 0
odd_count = 0

for num in numbers:
    if num % 2 == 0:
        even_count += 1
    else:
        odd_count += 1

print("Number of even elements:", even_count)
print("Number of odd elements:", odd_count)
```
